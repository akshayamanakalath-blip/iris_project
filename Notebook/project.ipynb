```python
import pandas as pd
```

```python
df=pd.read_csv("iris.csv")
```

```python
df.head()
```

```python
# checking missing values
```

```python
df.isna().sum()
```

```python
#separate input and output feature
```

```python
X=df.drop(labels=['Id','Species'],axis=1)
Y=df[['Species']]
```

```python
# preprocessing
```

```python
# X has only numeric features
```

```python
# first try this method
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler().set_output(transform='pandas')
X_pre=scaler.fit_transform(X)
```

```python
import pickle
with open('scaler.pkl','wb') as file1: 
    pickle.dump(scaler,file1)
```

```python
# check a file is added in notebook as scaler.pkl
```

```python
# divide 
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X_pre,Y,test_size=0.33,random_state=21)
```

```python
# model building
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
LogisticRegression()
```

```python
# Model Evaluation
```

```python
model.score(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

```python
# prediction of training and testing
```

```python
tr_pred=model.predict(xtrain)
ts_pred=model.predict(xtest)
```
